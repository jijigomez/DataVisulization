```python
import pandas as pd

# Load CSV into DataFrame
df = pd.read_csv('steals.csv')
```

```python
df
```

```python
df=pd.DataFrame(df)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
    'Unnamed: 0': 'Index',
    'GP Games played': 'GP',
    'GS Games started': 'GS',
    'AST Total Assists': 'AST',
    'APG Assists Per Game': 'APG',
    'TO Turnovers': 'TO',
    'TOPG Turnovers Per Game': 'TOPG',
    'A/TO Assists Per Turnover': 'A_TO',
    'STL Total Steals': 'STL',
    'SPG Steals Per Game': 'SPG',
    'Player': 'Player',
    'Position': 'Position',
    'Team': 'Team'
})
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
df.to_csv('preprocessed_basketball_stats.csv', index=False)
```

```python
avg_APG = df['APG  Assists Per Game'].mean()#calculate average assistants
print(f'Average APG: {avg_APG}')
```

```python
# Filter players who have started at least 10 games
df_filtered = df[df['GS  Games started'] >= 10]
```

```python
# Sort by total assists (AST) in descending order
df_sorted = df.sort_values(by='AST  Total Assists', ascending=False)
```

```python
df_sorted
```

```python
# Group by team and calculate average assists per game
avg_APG_by_team = df.groupby('Team')['APG  Assists Per Game'].mean()
print(avg_APG_by_team)
```

```python
import matplotlib.pyplot as plt

avg_APG_by_team.plot(kind='bar', figsize=(10, 6))
plt.title('Average Assists Per Game by Team')
plt.xlabel('Team')
plt.ylabel('Average APG')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting bar charts
plt.figure(figsize=(16, 12))

# Plot 1: Games Played (GP Games played)
plt.subplot(2, 2, 1)
plt.bar(df['Player'], df['GP  Games played'], color='blue')
plt.xlabel('Player')
plt.ylabel('Games Played')
plt.title('Games Played by Player')

# Plot 2: Assists Per Game (APG Assists Per Game)
plt.subplot(2, 2, 2)
plt.bar(df['Player'], df['APG  Assists Per Game'], color='green')
plt.xlabel('Player')
plt.ylabel('Assists Per Game')
plt.title('Assists Per Game by Player')

# Plot 3: Turnovers Per Game (TOPG Turnovers Per Game)
plt.subplot(2, 2, 3)
plt.bar(df['Player'], df['TOPG  Turnovers Per Game'], color='orange')
plt.xlabel('Player')
plt.ylabel('Turnovers Per Game')
plt.title('Turnovers Per Game by Player')

# Plot 4: Steals Per Game (SPG Steals Per Game)
plt.subplot(2, 2, 4)
plt.bar(df['Player'], df['SPG  Steals Per Game'], color='red')
plt.xlabel('Player')
plt.ylabel('Steals Per Game')
plt.title('Steals Per Game by Player')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(df['GP  Games played'], df['APG  Assists Per Game'], color='blue', alpha=0.5)
plt.xlabel('Games Played')
plt.ylabel('Assists Per Game')
plt.title('Scatter Plot: Games Played vs Assists Per Game')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

position_counts = df['Position'].value_counts()
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart: Distribution of Positions')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.hist(df['TOPG  Turnovers Per Game'], bins=10, color='orange', edgecolor='black')
plt.xlabel('Turnovers Per Game')
plt.ylabel('Frequency')
plt.title('Histogram: Turnovers Per Game')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df['Player'], df['SPG  Steals Per Game'], marker='o', linestyle='-', color='green')
plt.xlabel('Player')
plt.ylabel('Steals Per Game')
plt.title('Line Plot: Steals Per Game by Player')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.tight_layout()
plt.show()
```
